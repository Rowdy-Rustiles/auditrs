```python
"""

In this cell we open our example log and parse out the different type values for each field

"""
with open("../example_logs/redhat.log", 'r', encoding='utf-8') as f:
        content = f.readlines()
        types = set()
        for line in content:
            types.add(line[line.index("=") + 1:line.index(" ")])

        for type in types:
            type = type.split("_")
            print("".join([word[0] + word[1::].lower() for word in type]) +",")
```

```python
unique_field_combinations = set()

with open("../example_logs/redhat.log", "r", encoding="utf-8") as f:
    for raw_line in f:
        line = (
            raw_line
            .replace("\u001d", " ")
            .replace("{ ", "{")
            .replace(" }", "}")
            .strip()
        )

        if not line:
            continue

        type_name = line[line.index("=") + 1 : line.index(" ")]

        field_names = set()

        for part in line[line.index(" ") :].split():
            if "=" in part:
                key, _ = part.split("=", 1)
                field_names.add(key)
            elif part == ":":
                continue
            else:
                raise ValueError(f"Invalid line: {line}")

        # (type, field-set) uniquely identifies a combination
        unique_field_combinations.add(
            (type_name, frozenset(field_names))
        )

# Pretty-print
for type_name, fields in sorted(unique_field_combinations):
    print(f"{type_name}: {sorted(fields)}")
```
